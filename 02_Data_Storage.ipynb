```python
import pymysql
import csv
import time

# Database configuration
HOST = "localhost"
USER = "root"
PASSWORD = "Glitch@142"  # Replace with your MySQL password
DATABASE = "sales_dashboard"  # Replace with your database name
TABLE_NAME = "transactions"   # Replace with your actual table name
CSV_FILE = "sales_data_india.csv"

# Connect to MySQL using pymysql
connection = pymysql.connect(
    host=HOST,
    user=USER,
    password=PASSWORD,
    database=DATABASE,
    autocommit=True
)
cursor = connection.cursor()

# Create table if not exists (adjust data types as needed)
cursor.execute(f"""
CREATE TABLE IF NOT EXISTS {TABLE_NAME} (
    customer_id INT,
    customer_name VARCHAR(255),
    customer_age INT,
    customer_gender VARCHAR(10),
    customer_location VARCHAR(100),
    product_id INT,
    product_name VARCHAR(255),
    product_category VARCHAR(100),
    product_brand VARCHAR(100),
    product_rating FLOAT,
    transaction_id BIGINT PRIMARY KEY,
    quantity INT,
    price FLOAT,
    discount_applied FLOAT,
    total_amount FLOAT,
    payment_method VARCHAR(50),
    order_date DATE,
    order_time TIME,
    delivery_time DATETIME
)
""")

print("⏳ Starting stream from CSV to MySQL...")

while True:
    with open(CSV_FILE, "r", encoding='utf-8') as file:
        reader = csv.DictReader(file)
        new_rows = []
        for row in reader:
            # Prepare data tuple for insertion (insert all rows without filtering)
            new_rows.append((
                int(row['customer_id']),
                row['customer_name'],
                int(row['customer_age']),
                row['customer_gender'],
                row['customer_location'],
                int(row['product_id']),
                row['product_name'],
                row['product_category'],
                row['product_brand'],
                float(row['product_rating']),
                int(row['transaction_id']),
                int(row['quantity']),
                float(row['price']),
                float(row['discount_applied']),
                float(row['total_amount']),
                row['payment_method'],
                row['order_date'],
                row['order_time'],
                row['delivery_time']
            ))

        if new_rows:
            insert_query = f"""
            INSERT IGNORE INTO {TABLE_NAME} (
                customer_id, customer_name, customer_age, customer_gender, customer_location,
                product_id, product_name, product_category, product_brand, product_rating,
                transaction_id, quantity, price, discount_applied, total_amount, payment_method,
                order_date, order_time, delivery_time
            ) VALUES (
                %s, %s, %s, %s, %s,
                %s, %s, %s, %s, %s,
                %s, %s, %s, %s, %s, %s,
                %s, %s, %s
            )
            """
            cursor.executemany(insert_query, new_rows)
            print(f"✅ Attempted to insert {len(new_rows)} rows (duplicates ignored).")

        else:
            print("⏸ No data found in CSV.")

    time.sleep(10)  # Wait before next check
```
